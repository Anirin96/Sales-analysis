```python
import pandas as pd
import numpy as numpy
from datetime import datetime, timedelta
```

```python
sales_data = pd.read_csv('sales_data.csv')
sales_data
```

```python
products_data = pd.read_csv('products_data.csv')
products_data
```

```python
sales_data['order_date'] = pd.to_datetime(sales_data['order_date'])
sales_data
```

```python
merge_data = sales_data.merge(products_data, on='product_id')
merge_data
```

1. В каком регионе было продано наибольшее количество продуктов категории "Смартфоны"?

```python
smartphone = merge_data[merge_data['category'] == 'Смартфоны']
```

```python
region_with_most_smartphone_sales = smartphone.groupby('region')['quantity'].sum().idxmax()
region_with_most_smartphone_sales
```

2. Какова разница между максимальной и минимальной средней ценой продуктов категории "Ноутбуки" между всеми производителями? Ответ округлите до целых.

```python
laptops = merge_data[merge_data['category'] == 'Ноутбуки']
```

```python
manufacturer_avg_price = laptops.groupby('manufacturer')['price_per_unit'].mean()
round(max(manufacturer_avg_price) - min(manufacturer_avg_price))
```

3. Какой покупатель совершил заказы на наибольшую сумму по всем временам?

```python
customer_total_spending = sales_data.groupby('customer_id')['total_price'].sum()
customer_total_spending
```

```python
top_customer = customer_total_spending.idxmax()
top_customer
```

4. Какова доля продаж продуктов категории "Планшеты" от общей выручки за все время? Ответ округлите до 2 знака после запятой. Разделитель разряда - точка.

```python
tablets_data = merge_data[merge_data['category'] == 'Планшеты']
```

```python
taplets_revenue = tablets_data['total_price'].sum() / merge_data['total_price'].sum()
round(taplets_revenue, ndigits=2)
```

5. Какова разница в средней цене продуктов категории "Смартфоны" между двумя самыми крупными регионами по продажам? Ответ округлите до целых.

```python
region_revenue = merge_data.groupby('region')['total_price'].sum()
```

```python
top2_regions = region_revenue.nlargest(2)
top2_regions
```

```python
top_regions_smatrphones = merge_data[(merge_data['region'].isin(top2_regions.index)) & (merge_data['category'] == 'Смартфоны')]
avg_price_by_region = top_regions_smatrphones.groupby('region')['price_per_unit'].mean()
price_difference = avg_price_by_region.diff().abs().iloc[-1]
round(price_difference)
```

6. Сколько в среднем уникальных категорий продуктов было продано по всем регионам?

```python
unique_categories = merge_data.groupby('region')['category'].nunique()
round(unique_categories.mean())
```

8. В каком месяце было продано больше всего продуктов категории "Планшеты" в регионе "Север"?

```python
north_tablets = merge_data[(merge_data['category'] == 'Планшеты') & (merge_data['region'] == 'Север')]
north_tablets['month'] = north_tablets['order_date'].dt.month
north_tablets.groupby('month')['quantity'].sum().idxmax()
```

9. Какой производитель продал продукты на наибольшую сумму в регионе "Юг" за март 2023 года?

```python
south_march_2023 = merge_data[(merge_data['region'] == 'Юг') & (merge_data['order_date'].dt.month == 3) & (merge_data['order_date'].dt.year == 2023)]
manufacturer_revenue_south_march_2023 = south_march_2023.groupby('manufacturer')['total_price'].sum()
manufacturer_highest_revenue = manufacturer_revenue_south_march_2023.idxmax()
manufacturer_highest_revenue
```

10. Сколько покупателей совершили заказы на продукты хотя бы трех разных категорий?

```python
customer_unique_categories = merge_data.groupby('customer_id')['category'].nunique()
multi_cus = customer_unique_categories[customer_unique_categories >= 3]
number_multi_cus = len(multi_cus)
number_multi_cus
```

11. Какова доля продаж продуктов производителя "C" от общей выручки за все время в регионе "Север"? Ответ округлите до 2 знака после запятой. В качестве разделителя разрядов используйте точку.

```python
c_north = merge_data[(merge_data['manufacturer'] == 'C') & (merge_data['region'] == 'Север')]
revenue_ratio = c_north['total_price'].sum() / merge_data[(merge_data['region'] == 'Север')]['total_price'].sum()
round(revenue_ratio, ndigits=2)
```

12. Найдите покупателей, которые купили у более чем 15% уникальных производителей. Если таких покупателей несколько - напишите их через запятую с пробелом в порядке возрастания.

Примечание: Уникальными производителями считайте уникальную связку "Категория = Производитель".

```python
total_manufactures = int(merge_data.groupby(['category', 'manufacturer'])['manufacturer'].agg({'nunique'}).sum())
unique_manufacturer_per_customer = merge_data.groupby(['customer_id', 'category', 'manufacturer'])['manufacturer'].agg({'nunique'})
percent_from_total = unique_manufacturer_per_customer.groupby('customer_id').sum().apply(lambda x: x / total_manufactures).reset_index()
customers_with_more_than15 = percent_from_total[percent_from_total['nunique'] > 0.15]
customers_with_more_than15['customer_id'].tolist()
```

14. Какой производитель имеет самую высокую среднюю цену продукта среди всех категорий?

```python
average_price_per_manufacturer = merge_data.groupby(['manufacturer', 'category'])['price_per_unit'].mean()
manufacturer_with_highest_average_price = average_price_per_manufacturer.groupby('manufacturer').mean().idxmax()
manufacturer_with_highest_average_price
```

15. Каково общее количество уникальных производителей, чьи продукты были куплены менее чем в 3 регионах?

```python
unique_regions_per_manufacturer = merge_data.groupby(['manufacturer', 'region'])['region'].nunique()
manufacturers_with_products_in_multiple_regions = unique_regions_per_manufacturer.groupby('manufacturer').apply(lambda x: (len(x) <= 2))
total = manufacturers_with_products_in_multiple_regions.sum()
total
```
